```python
from sklearn.preprocessing import StandardScaler
scaling=StandardScaler()
y=['season','yr','mnth','holiday','weekday','workingday','weathersit','casual','registered','cnt']
data1=scaling.fit_transform(data.drop(y,axis=1))
```

```python
data.columns
```

```python
# convert array into data frame
data2=pd.DataFrame(data1,columns=['temp', 'atemp', 'hum', 'windspeed', ])
```

```python
#create data frame for discreate data
df1=data[['season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday',
       'weathersit', 'casual','registered', 'cnt']]
    
```

```python
new_data=pd.concat([df1,data2],axis=1)
```

# feature engineering

```python
new_data.corr()
```

```python
new_data.drop(["holiday"],inplace=True,axis=1)#drop the irrelavant column
```

# model building

```python
#split into x and y
x = new_data.drop('cnt',axis=1)
y = new_data['cnt']
```

```python
#splitting the entire data into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.25,random_state=32)
```

```python
new_data.shape
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
# implement the model

# import the algorithum
from sklearn.linear_model import LinearRegression

# initializatin
model = LinearRegression()

# train the model
model.fit(x_train,y_train)
```

```python
#test the model / getting prediction
y_pred = model.predict(x_test)
y_pred
```

```python
# slopes # 3 slope values are given? 
# m1 x1 + m2 x2 + m3 x3 + C = y # since we have 3 input values 3 slopes#
model.coef_
```

```python
# Erros / loss functions 
# Error = Actual - predicted value
# MSE
# MAE 
# RMSE
# takes 2 parameters y_test, y_pred
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mse = mean_squared_error(y_test,y_pred)
mse
```

```python
mae = mean_absolute_error(y_test,y_pred)
mae
```

```python
rmse = np.sqrt(mse)
rmse
```

```python
r2_score(y_test,y_pred)
```

```python
#Conclusion
# 97% is the r2 score
# this signifies, the over all perfomance of the model is good
# 97% of the records are predicted correctly
# 3% of data is predicted wrong
```
